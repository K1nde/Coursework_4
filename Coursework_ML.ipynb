# 1. Описание данных

```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt

telecom = pd.read_csv('telecom_users.csv')
telecom.head()
```

```python
# В датасете есть признаки, которые никак не повлияют на предсказания:

# customerID, PaymentMethod, InternetService
# поэтому от них можно избавиться.
# Так же удаляем столбец 'Unnamed: 0'
```

```python
telecom = telecom.drop(['Unnamed: 0', 'customerID', 'PaymentMethod', 'InternetService', 'MonthlyCharges', 'TotalCharges'], axis=1)
```

```python
telecom.info()
```

```python
# пропущенных значений в датасете нет, но все значения нужно перевести в булевые значения
```

```python
telecom = telecom.replace('Yes', 1)
telecom = telecom.replace('No', 0)
telecom = telecom.replace('No internet service', 0)
telecom = telecom.replace('No phone service', 0)
telecom = telecom.replace(' ', 0)
telecom['gender'] = telecom['gender'].map( {'Female': 0, 'Male': 1} ).astype(int)
telecom['Contract'] = telecom['Contract'].map( {'Month-to-month': 0, 'One year': 1,'Two year': 2} ).astype(int)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
telecom.hist(bins = 50, figsize= (20,15))
plt.show()
```

```python
colormap = plt.cm.RdBu
plt.figure(figsize=(14,12))
plt.title('Correlation of Features', y=1.05, size=15)
sns.heatmap(telecom.astype('float64').corr(),linewidths=0.1,vmax=1.0, 
            square=True, cmap=colormap, linecolor='white', annot=True)
```

# 2. Исследование зависимостей и формулирование гипотез

```python
# Признаки OnlineSecurity, OnlineBackup, DeviceProtection, TechSupport, StreamingTV, StreamingMovies имеют 
# зависимость между собой, создадим новый признак 'Service', который будет показывать количество используемых сервисов.
```

```python
telecom['Service'] = (telecom['OnlineSecurity'] + 
                      telecom['OnlineBackup'] + 
                      telecom['DeviceProtection'] +
                      telecom['TechSupport'] + 
                      telecom['StreamingTV'] + 
                      telecom['StreamingMovies'])
```

```python
plt.figure(figsize= (20,10))
sns.countplot(x = telecom['Service'], hue = telecom['Churn'])
plt.show()
```

```python
telecom = telecom.drop(['OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies'], axis=1)
```

# 3. Построение моделей для прогнозирования оттока

```python
# Были взяты 6 моделей классификации для прогнозирования оттока
```

```python
from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV

from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier
from xgboost import XGBClassifier



x_train, x_test, y_train, y_test = train_test_split(telecom.loc[:, telecom.columns != 'Churn'],
                                                    telecom.Churn,
                                                    test_size=0.3,
                                                    random_state=17,
                                                    stratify=telecom.Churn)
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
dtc = DecisionTreeClassifier()
lr = LogisticRegression(class_weight={0:2, 1:8})
rfc = RandomForestClassifier()
gbc = GradientBoostingClassifier()
xgb = XGBClassifier(eval_metric='logloss')

models = [knn, dtc, lr, rfc, gbc, xgb]
```

```python
# С помощью GridSearchCV найдем лучшие параметры для моделей
```

```python
knn_patams = {'n_neighbors': range(1, 11, 1)}

dtc_patams = {'min_samples_leaf': range(1, 5)}

lr_params = {'C': np.arange(0.5, 1, 0.1)}

rfc_params = {'n_estimators': range(10, 100, 10), 
              'min_samples_leaf': range(1, 5)}

gbc_params = {'learning_rate': np.arange(0.1, 0.6, 0.1)} 

weights = [1, 10, 25, 50, 75, 99, 100, 1000]
xgb_params = {'eta': np.arange(0.1, 1, 0.1),
              'max_depth': range(0, 5, 1),
              'scale_pos_weight': weights}



skf = StratifiedKFold(n_splits=3, random_state=17, shuffle=True)

grid_params = [knn_patams, dtc_patams, lr_params, rfc_params, gbc_params, xgb_params]
```

```python
grid_models = []
for n,model in enumerate (models):
    model = GridSearchCV(model, grid_params[n], cv=skf)  
    grid_models.append(model)
```

# 4. Сравнение качества моделей 

```python
# Мы будем используем ROC-AUC в связи с тем, что набор данных у вас несбалансирован.
```

```python
from sklearn.metrics import roc_auc_score
```

```python
best_score = 0
best_model = []
for model in (grid_models):
    model.fit(x_train, y_train)
    probas = model.predict_proba(x_test)
    score = roc_auc_score(y_test, probas[:, 1])
    print(model,'roc_auc_score =', score)
    if score > best_score:
        best_score = score
        best_model = model
```

```python
print('\nBest model - ',best_model, best_score)
```
